```python
todo_list = []

def show_menu():
    print("\n====== To-Do List Menu ======")
    print("1. View Tasks")
    print("2. Add Task")
    print("3. Mark Task as Complete")
    print("4. Delete Task")
    print("5. Exit")

while True:
    show_menu()
    choice = input("Enter your choice (1-5): ")

    if choice == '1':
        if len(todo_list) == 0:
            print("\nNo tasks in your to-do list.")
        else:
            print("\nYour Tasks:")
            for i, task in enumerate(todo_list):
                status = "Done" if task['completed'] else "Not Done"
                print(f"{i + 1}. {task['name']} - [{status}]")

    elif choice == '2':
        task_name = input("Enter the task name: ")
        task = {'name': task_name, 'completed': False}
        todo_list.append(task)
        print("Task added!")

    elif choice == '3':
        if len(todo_list) == 0:
            print("No tasks to mark as complete.")
        else:
            task_num = input("Enter the task number to mark as complete: ")
            if task_num.isdigit() and 1 <= int(task_num) <= len(todo_list):
                todo_list[int(task_num) - 1]['completed'] = True
                print("Task marked as complete!")
            else:
                print("Invalid task number.")

    elif choice == '4':
        if len(todo_list) == 0:
            print("No tasks to delete.")
        else:
            task_num = input("Enter the task number to delete: ")
            if task_num.isdigit() and 1 <= int(task_num) <= len(todo_list):
                deleted_task = todo_list.pop(int(task_num) - 1)
                print(f"Deleted task: {deleted_task['name']}")
            else:
                print("Invalid task number.")

    elif choice == '5':
        print("Goodbye!")
        break

    else:
        print("Invalid choice. Please enter a number from 1 to 5.")
```
